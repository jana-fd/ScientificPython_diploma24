#### **Exercise 1:**

Repeat the curve fitting example with the following modifications:

 - choose some other function
 - first write the data to a file, then read it (with numpy, for example)
 - produce the picture that has axes, name and legend

```python
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)

# Let's generate fake data that we can "fit" then 
def f(x, a, phi):
    return np.sin(a*x + phi) * np.exp(-1.0*x)

#random makes it look like experimental data
xp = np.linspace(0, 4, 50)
yp = f(xp, 1.3, 0.9) + .1*np.random.normal(size=50)

xyp = np.concatenate((xp,yp), axis=0)


np.savetxt('my_function.txt', xyp)


xy = np.loadtxt('my_function.txt')

x, y = np.split(xy, 2) 

# Now fit the model: the parameters omega and phi can be found in the
# `params` vector

params, params_cov = optimize.curve_fit(f, x, y)

print(params)

# plot the data and the fitted curve
plt.plot(x, y, 'bx')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,f(x,1.3,0.9), 'g-')
plt.plot(x, f(x, *params), 'r-')
plt.title('sin(1.3x + 0.9) * exp(-x)')
plt.legend('curve')
```

#### **Exercise 2:**

Integrate the function `x**2 + 10*np.sin(x)` on the interval `[-10,10]`.
Find a way to make a plot like in textbooks where you color the area under the curve that you integrate.
Now integrate the same function symbolically using `sympy`, then put borders into it and compare the results.

```python
from scipy.integrate import quad

def func(x):
    return x**2 + 10*np.sin(x)

x = np.linspace(-10,10,50)
y = func(x)
result, error = quad(func, -10, 10)
plt.plot(x, y, label='The curve', color = 'black')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')

plt.fill_between(x, y, 0, color= 'cornflowerblue')
```

#### **Exercise 3**

With the help of documentation find 3 different ways to interpolate data produced from the function 
`np.cos(-x**2/9.0)`. 
Create a class called "My_Interpolation" with these 3 interpolation methods.

```python
from scipy.interpolate import interp1d
from scipy.interpolate import splrep, splev
from scipy.interpolate import Rbf

class My_Interpolation:
    def __init__(self,x,y):
        self.x = x
        self.y = y

    def linear_interpolation(self, x):
        f = interp1d(self.x, self.y, kind='cubic')
        return f(x)

    def spline_interpolation(self, x):
        sp = splrep(self.x, self.y)
        f = splev(x, sp)
        return f

    def radialBasis_interpolation(self, x):
        f = Rbf(self.x, self.y)
        return f(x)

x = np.linspace(0, 10, 17)
y = np.cos(-x**2/9.0)
plt.plot(x, y, 'o',label= 'original values')
interpolated = My_Interpolation(x,y)
xnew = np.linspace(0, 10, 100)
ynew = interpolated.linear_interpolation(xnew)
plt.plot(xnew, ynew, label='Linear Interpolation')

ynew1 = interpolated.spline_interpolation(xnew)
plt.plot(xnew, ynew1, label='Spline Interpolation')

ynew2 = interpolated.radialBasis_interpolation(xnew)
plt.plot(xnew, ynew2, label='Radial Basis Function')
plt.legend()
    
```

#### **Exercise 4:**
Give students in the `example.csv` file a bonus of 5 for all the courses. Find a way to replace all the values that are now greater than 30 by 30. Save the new data to a file. Make a plot with histograms for every course in the example file.

```python
import pandas as pd
df = pd.read_csv('data/example.csv', index_col=0)
df = df + 5
df[df > 30] = 30
print(df) 

df.to_csv('new_example.csv', index=False)

fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize=(10,8))
df["course1"].plot(kind='hist', ax=axes[0,0])
df["course2"].plot(kind='hist', ax=axes[0,1])
df["course3"].plot(kind='hist', ax=axes[1,0])
df["course4"].plot(kind='hist', ax=axes[1,1])
plt.tight_layout()
```

#### **Exercise 5:**

 - Read in the `cars.csv` file as a dataframe
 - Remove the row with the index 3, which contains a quote in the Model column.
 - Add a new column called Price with the following values: [8000, 6500, 7000, 6800, 7500, 7300, 7000, 9000, 6500, 7800]
 - Find the average Length of all the vehicles.
 - Find the median Price of all the vehicles.
 - Create a new dataframe that only contains the rows where the Price is greater than 7000.

```python
df = pd.read_csv('data/cars.csv', index_col=0)

prices = [8000, 6500, 7000, 6800, 7500, 7300, 7000, 9000, 6500, 7800]
df['Price'] = prices
df1 = df.drop([1993], axis='rows')
print(df1)
print('average length = ' + str(df1["Length"].mean()))
print('average price = ' + str(df1["Price"].mean()))

df2 = df[df['Price'] >= 7000]
print(df2)

df2 = df2.sort_values(by='Price', ascending=False)
print(df2)
```

# **Exercise 6:**
 - create a random matrix with numpy
 - save it to file
 - read it back with numpy
 - calculate some statistcs on it with numpy 
 - read the same file with pandas
 - calculate the same statistics on it with pandas

```python
N = 5
M = 4

random_matrix = np.random.rand(N,M)
print(random_matrix)

np.savetxt('randomMatrix.txt', random_matrix)

mat = np.loadtxt('randomMatrix.txt')
mean1 = np.mean(mat)
print('mean using numpy: ' + str(mean1))
med1 = np.median(mat)
print('median using numpy: ' +str(med1))
std_dev1 = np.std(mat)
print('standard deviation using numpy: ' +str(std_dev1))

mat1 = pd.read_csv('randomMatrix.txt', delimiter =' ', header=None)
print(mat1)

flat = mat1.values.flatten()
mean2 = flat.mean()
print('mean using pandas: ' + str(mean2))
med2 = pd.Series(flat).median()
print('median using pandas: ' + str(med2))
std_dev2 = pd.Series(flat).std(ddof = 0)
print('standard deviation using pandas: ' +str(std_dev2))
```
