### **Problem1:**

Write a function that calculates the first N [Fibonacci numbers](https://en.wikipedia.org/wiki/Fibonacci_number).

````python
def fibonacci(N):
    fib = list()
    if N < 2:
        result = N
        fib.append(result)
    else:
        result = fibonacci(N-1) + fibonacci(N-2)
        fib.append(result)
    return result

N = 12
print(fibonacci(N))
````

### **Problem2:**

Generate random data for 2D points coordinates. Write a function that plots them with different colors for each 90 degree section with the origin in the center. This problem should be coded as a class, i.e. you should be able to execute the code:

```
pr=Problem2()
pr.generate()
pr.display()
```

````python
class Problem2:
    def __init__(self, pointsNum = 50):
        self.pointsNum = pointsNum
        self.pointsCartesian = None
        self.pointsPolar = None
        
    def generate(self):
        x = np.random.uniform(-10,10,self.pointsNum)
        y = np.random.uniform(-20,20,self.pointsNum)
        self.pointsCartesian = np.column_stack((x,y))

        #calculate polar coordinates
        angles = np.arctan2(y, x)
        distances = np.sqrt(x**2 + y**2)
        self.pointsPolar = np.column_stack((angles, distances))

    def display(self):
        angles = self.pointsPolar[:, 0]
        colors = (np.degrees(angles) + 360) % 360
        quadrants = np.zeros(self.pointsNum, dtype = int)
        quadrants[(colors >= 0) & (colors < 90)] = 1
        quadrants[(colors >=90) & (colors < 180)] = 2
        quadrants[(colors >= 180) & (colors < 270)] = 3
        quadrants[(colors >=270) & (colors < 360)] = 4
        plt.scatter(self.pointsCartesian[:, 0], self.pointsCartesian[:, 1], c=quadrants)
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Random Points Generator')
        plt.show()


N = 200
pr = Problem2(N)
pr.generate()
pr.display()
````

### **Problem3:**

Write a decorator that calls a function twice.

````python
def twice(f, *args, **kwargs):
    def wrapper(*args, **kwargs):
        x = f(*args, **kwargs)
        y = f(x)
        return x, y
    return wrapper

@twice
def f(x):
    return x*x*x - 2*x*x + 3*x - 1

x, y = f(-2)
print('first function call: ', x)
print('second function call: ', y)
    
````

### **Problem4:**

Use sympy to solve the equation `d^2y(x)/dx^2 + 9y(x)+x=0`
Check that it's correct (still using sympy). Hint: you can access the right hand side of solution expression with `.rhs`.

Then set the constants equal to 1 and integrate the obtained function numerically using scipy over the interval [0,10]
Hint: you can use symply's lambdify function to turn sympy's expression into a function you can use later for scipy.


````python
import sympy as sp
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

y = sp.Function('y')
x = sp.symbols('x')
eq = sp.Eq(y(x).diff(x,x) + 9*y(x) + x, 0)
sol = sp.dsolve(eq, y(x))
print(sol)
rhs_sol = sol.rhs
rhs_sol = rhs_sol.subs({sp.symbols('C1'): 1, sp.symbols('C2'): 1})
print(rhs_sol)
lhs = eq.lhs.subs(y(x), rhs_sol).simplify()
rhs = eq.rhs.subs(y(x), rhs_sol).simplify()
print(lhs == rhs)


y_numeric = sp.lambdify(x, rhs_sol, 'numpy')
x_values = np.linspace(0,10,100)
initial_conditions = [0,1]
result = odeint(lambda y, x:[y[1], -9*y[0] -x], initial_conditions, x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, result[:, 0], label='Integration')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Integration of the equation')
plt.legend()
plt.show()
````

### **Problem5:**
(advanced, would be given as a project to do at home and explain at the oral exam)
**Solving Laplace equation by Jacobi method**

The algorithm used in this problem solves Laplaces equation on an evenly spaced grid through the use of a simple Jacobi iteration technique. The equation is a partial differential equation named after Pierre-Simon Laplace and are important in many fields of science: namely the fields of electromagnetism, astronomy and fluid dynamics. 

The equation has the form: 

```
d^2f/dx^2 + d^2f/dy^2=0
```

**The Algorithm (implement this as a class)**

 - create a numpy array to store 2 dimensional grid.

 -  Setup the initial constant boundary conditions in the following way:
     - the value at the lower left hand corner of the of the grid is 100, 
     - the value ascending and to the right is set to a linear gradient reaching zero at the opposite corners (see Figure1). The rest of the borders are fixed at zero. Please note, that these boundaries remain constant throughout the simulation.

 - Setup the initial condition of the inner grid elements as 0.5.

-  Write a jacobi method. At each iteration, the value of each inner matrix element needs to be recomputed from elements of the current iteration (in a similar way to how we did with game of life). The updating formula, based on numerical computation of second derivatives, is:

```
new_field[i,j]=0.25*(field[i,j-1]+field[i,j+1]+field[i+1,j]+field[i-1,j])
```

![](jacobiFigure1.jpg)

Use `imshow` to plot the result. There is a `jacobi.gif` file uploaded that shows how the solution looks like. Making gifs in python is troublesome, so just the final picuture would be enough.

````python
import numpy as np
import matplotlib.pyplot as plt
class Jacobi:
    def __init__(self, xsize, ysize, iterations=500, tolerance = 10e-4):
        self.xsize = xsize
        self.ysize = ysize
        self.iterations = iterations
        self.tolerance = tolerance
        self.arr2d = np.zeros((xsize, ysize), dtype = float)

    def initial_conditions(self):
        self.arr2d[-1, 0] = 100
        self.arr2d[:, 0] = np.linspace(0, 100, self.xsize)
        self.arr2d[-1, :] = np.linspace(100,0, self.ysize)
        self.arr2d[0, :] = 0.0
        self.arr2d[:, -1] = 0.0
        self.arr2d[1:-1, 1:-1] = 0.5
        

    def display(self):
         plt.imshow(self.arr2d, cmap='bwr', origin='upper', extent=[0, self.ysize, 0, self.xsize])
         plt.colorbar()
         plt.title("Laplace Equation - Jacobi Method")
         plt.show()

    def jacobi_method(self):
        for iteration in range(self.iterations):
            new_arr = np.copy(self.arr2d)
            for i in range(1, self.xsize - 1):
                for j in range(1, self.ysize - 1):
                    new_arr[i, j] = 0.25 * (self.arr2d[i, j-1] + self.arr2d[i, j+1] + self.arr2d[i+1, j] + self.arr2d[i-1, j])
            diff = np.max(np.abs(new_arr - self.arr2d))
            self.arr2d = new_arr
            if diff < self.tolerance:
                print("Converged after " + str(iteration+1) + " iterations")
                break
    

arr = Jacobi(10,10,1000)
arr.initial_conditions()
arr.display()
arr.jacobi_method()
arr.display()

        
        
    
````
